```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Окружность

```python
x1 = []
x2 = []
y1 = []
y2 = []
t = 120 # число точек
r = 1
for i in range(t):
    x_ = i/(t-1)
    x1.append(x_)
    x2.append(-x_)
    y1.append(math.sqrt(r**2 - x_**2))
    y2.append(-math.sqrt(r**2 - x_**2))
plt.plot(x1, y1)
plt.plot(x2, y1)
plt.plot(x1, y2)
plt.plot(x2, y2)
plt.xlabel("x")
plt.ylabel("y")
plt.axis('scaled')
```

# Эллипс

```python
x1 = []
x2 = []
y1 = []
y2 = []
t = 120 # число точек
a = 15
b = 4
for i in range(t):
    x_ = i/((t-1)/a)
    x1.append(x_)
    x2.append(-x_)
    y1.append(math.sqrt((b**2)*(1 - x_**2/a**2)))
    y2.append(-math.sqrt((b**2)*(1 - x_**2/a**2)))
plt.plot(x1, y1)
plt.plot(x2, y1)
plt.plot(x1, y2)
plt.plot(x2, y2)
plt.xlabel("x")
plt.ylabel("y")
plt.axis('scaled')
```

# Гипербола

```python
x1 = []
x2 = []
y1 = []
y2 = []
t = 120 # число точек
a = 20
b = 15
for i in range(t):
    x_ = i+a
    x1.append(x_/10)
    x2.append(-x_/10)
    y1.append(math.sqrt((b**2)*(x_**2/a**2 - 1))/10)
    y2.append(-math.sqrt((b**2)*(x_**2/a**2 - 1))/10)
plt.plot(x1, y1)
plt.plot(x2, y1)
plt.plot(x1, y2)
plt.plot(x2, y2)
plt.xlabel("x")
plt.ylabel("y")
plt.axis('scaled')
```

# Нарисуйте трехмерный график двух параллельных плоскостей

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 2*X + 5*Y
Z1 = 2*X + 5*Y + 100
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z1)
ax.scatter(0, 0, 0, 'z', 50, 'red')
show
```

# Нарисуйте трехмерный график двух любых поверхностей второго порядка

# Эллипсоид

```python
fig = figure()
ax = Axes3D(fig)
a = 4
b = 4
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(1 - X**2/a**2 - Y**2/b**2)
Z1 = -np.sqrt(1 - X**2/a**2 - Y**2/b**2)
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z1)
ax.scatter(0, 0, 0, 'z', 50, 'red')
show
```

# Однополостный гиперболоид

```python
fig = figure()
ax = Axes3D(fig)
a = 2
b = 2
c = 3
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
Z = (c**2)*np.sqrt(X**2/a**2 + Y**2/b**2 - 1)
Z1 = -(c**2)*np.sqrt(X**2/a**2 + Y**2/b**2 - 1)
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z1)
ax.scatter(0, 0, 0, 'z', 50, 'red')
show
```
