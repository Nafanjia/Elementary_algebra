# Окружность

```python
x1 = []
x2 = []
y1 = []
y2 = []
t = 120 # число точек
r = 1
for i in range(t):
    x_ = i/(t-1)
    x1.append(x_)
    x2.append(-x_)
    y1.append(math.sqrt(r**2 - x_**2))
    y2.append(-math.sqrt(r**2 - x_**2))
plt.plot(x1, y1)
plt.plot(x2, y1)
plt.plot(x1, y2)
plt.plot(x2, y2)
plt.xlabel("x")
plt.ylabel("y")
plt.axis('scaled')
```

# Эллипс

```python
x1 = []
x2 = []
y1 = []
y2 = []
t = 120 # число точек
a = 15
b = 4
for i in range(t):
    x_ = i/((t-1)/a)
    x1.append(x_)
    x2.append(-x_)
    y1.append(math.sqrt((b**2)*(1 - x_**2/a**2)))
    y2.append(-math.sqrt((b**2)*(1 - x_**2/a**2)))
plt.plot(x1, y1)
plt.plot(x2, y1)
plt.plot(x1, y2)
plt.plot(x2, y2)
plt.xlabel("x")
plt.ylabel("y")
plt.axis('scaled')
```

# Гипербола

```python
x1 = []
x2 = []
y1 = []
y2 = []
t = 120 # число точек
a = 20
b = 15
for i in range(t):
    x_ = i+a
    x1.append(x_/10)
    x2.append(-x_/10)
    y1.append(math.sqrt((b**2)*(x_**2/a**2 - 1))/10)
    y2.append(-math.sqrt((b**2)*(x_**2/a**2 - 1))/10)
plt.plot(x1, y1)
plt.plot(x2, y1)
plt.plot(x1, y2)
plt.plot(x2, y2)
plt.xlabel("x")
plt.ylabel("y")
plt.axis('scaled')
```

# Решите систему уравнений:
y = x**2 – 1

exp(x) + x∙(1 – y) = 1

```python
x = np.linspace(-2, 5, 200)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2, 20)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)

x1, y1 = fsolve(equations, (-1, 1))
print(x1, y1)

x2, y2 = fsolve(equations, (2.5, 5))
print(x2, y2)

x3, y3 = fsolve(equations, (4, 17))
print(x3, y3)
```
