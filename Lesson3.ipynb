```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Окружность

```python
x1 = []
x2 = []
y1 = []
y2 = []
t = 120 # число точек
r = 1
for i in range(t):
    x_ = i/(t-1)
    x1.append(x_)
    x2.append(-x_)
    y1.append(math.sqrt(r**2-x_**2))
    y2.append(-math.sqrt(r**2-x_**2))
plt.plot(x1, y1)
plt.plot(x2, y1)
plt.plot(x1, y2)
plt.plot(x2, y2)
plt.xlabel("x")
plt.xlabel("y")
plt.axis('scaled')
```

# Эллипс

```python
x1 = []
x2 = []
y1 = []
y2 = []
t = 120 # число точек
a = 15
b = 4
for i in range(t):
    x_ = i/((t-1)/a)
    x1.append(x_)
    x2.append(-x_)
    y1.append(math.sqrt((b**2)*(1-x_**2/a**2)))
    y2.append(-math.sqrt((b**2)*(1-x_**2/a**2)))
plt.plot(x1, y1)
plt.plot(x2, y1)
plt.plot(x1, y2)
plt.plot(x2, y2)
plt.xlabel("x")
plt.xlabel("y")
plt.axis('scaled')
```

# Гипербола

```python
x1 = []
x2 = []
y1 = []
y2 = []
t = 120 # число точек
a = 20
b = 15
for i in range(t):
    x_ = i+a
    x1.append(x_/10)
    x2.append(-x_/10)
    y1.append(math.sqrt((b**2)*(x_**2/a**2-1))/10)
    y2.append(-math.sqrt((b**2)*(x_**2/a**2-1))/10)
plt.plot(x1, y1)
plt.plot(x2, y1)
plt.plot(x1, y2)
plt.plot(x2, y2)
plt.xlabel("x")
plt.xlabel("y")
plt.axis('scaled')
```
